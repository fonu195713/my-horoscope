```python
from selenium.webdriver.chrome.options import Options
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re
```

```python
options = Options()
options.add_argument("--disable-notifications")
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```python
chrome = webdriver.Chrome('chromedriver',options=options)
chrome.get("https://www.ximizi.net/24jieqi_1.php")

data = {}

for y in range(1899, 2101):
  chrome.implicitly_wait(2)
  try:
    WebDriverWait(chrome, 10, 0.5).until(EC.presence_of_element_located((By.CLASS_NAME, "ximizi_select")))
    WebDriverWait(chrome, 10, 0.5).until(EC.presence_of_element_located((By.CLASS_NAME, "ximizi_submit")))
  except Exception as e:
    print(e)
    exit()
  
  select_tag = Select(chrome.find_element(by=By.CLASS_NAME, value="ximizi_select"))
  select_tag.select_by_value(str(y))
  button = chrome.find_elements(by=By.CLASS_NAME, value="ximizi_submit")[0]
  button.click()
  chrome.implicitly_wait(2)
  
  div = chrome.find_elements(by=By.CLASS_NAME, value="f2_text")[0].text
  yearly_str = [match for match in re.findall('(公元.{17})', str(div))]

  for monthly_str in yearly_str[0::2]:
    month = (monthly_str.replace(" ",""))
    day = re.findall(r'\d+', month)
    day = [int(i) for i in day]

    if(day[0] == 1899 and day[1] == 1):
      continue
    
    key = "y"+str(day[0]) if(day[1] != 1) else "y"+str(day[0]-1)
    if(day[1] == 2):
      data[key] = []
    data[key].append(day)
```
